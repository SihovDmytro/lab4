```python
import os
import glob
import base64
import pandas as pd
from PIL import Image
from IPython.display import HTML

def get_image_info(directory):
    image_data = []

    for filename in os.listdir(directory):
        file_path = os.path.join(directory, filename)

        try:
            with Image.open(file_path) as img:
                image_name = filename
                color_model = img.mode
                image_format = img.format
                channels = len(img.getbands())
                size_mb = os.path.getsize(file_path) / (1024 * 1024.0)
                width, height = img.size
                full_path = file_path

                image_data.append({
                    'Image name': image_name,
                    'Color model': color_model,
                    'Format': image_format,
                    'Channels': channels,
                    'Image size': f'{size_mb:.2f}',
                    'Image width': width,
                    'Image height': height,
                    'Full path': full_path
                })
        except Exception as e:
            pass


    image_df = pd.DataFrame(image_data)

    return image_df

directory_path = "C:/Users/User/python/lab4"
get_image_info(directory_path)

```

```python
import os
import random
from PIL import Image, ImageOps, ImageFilter, ImageEnhance

def generate_poster(directory, output_name, columns, rows):
    image_files = [f for f in os.listdir(directory) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

    poster_width = columns * 500
    poster_height = rows * 500
    poster = Image.new('RGB', (poster_width, poster_height), color='white')

    for col in range(columns):
        for row in range(rows):
            image_file = random.choice(image_files)

            with Image.open(os.path.join(directory, image_file)) as img:
                img = ImageOps.exif_transpose(img)
                img = ImageOps.fit(img, (500, 500), method=0, bleed=0.0, centering=(0.5, 0.5))
                poster.paste(img, (col * 500, row * 500))

    poster.save(output_name)


generate_poster("C:/Users/User/python/lab4/task2", "poster1.jpg", 3, 2)
generate_poster("C:/Users/User/python/lab4/task2", "poster2.jpg", 4, 2)
generate_poster("C:/Users/User/python/lab4/task2", "poster3.jpg", 3, 3)
```
